```python
%%R
library(ggplot2)
library(plyr); library(dplyr)
library(tidyr)
```

```python
%%R -h 300
C.plot = function(x, x.peak = 0.5, c0 = 1, sigma = .05) {
    c0 * exp(-(x.peak - x)**2 / sigma ** 2)
}

d = data.frame(x = seq(0,1,0.001), C = sapply(seq(0,1,0.001), C.plot))

p = ggplot(d, aes(x = x, y = C))

p = p + geom_point()

p
```

```python
%%R -h 300
n = 30000
BD = c(rnorm(n, mean = 0.25, sd = 0.15), rnorm(n, mean = 0.95, sd = 0.15))
BD = (BD - min(BD)) / (max(BD) - min(BD))

OTU.1.BD = data.frame(BD = BD)
OTU.1.BD$OTU = "OTU.1"

binwidth = 0.05

p = ggplot(OTU.1.BD, aes(x = BD))

p = p + geom_histogram(alpha = 0.5, binwidth = binwidth, aes(y = ..density..))

p
```

```python
%%R -h 300
n = 30000
BD = c(rnorm(n, mean = 0.25, sd = 0.55), rnorm(n, mean = 0.55, sd = 0.025))
BD = (BD - min(BD)) / (max(BD) - min(BD))

OTU.2.BD = data.frame(BD = BD)
OTU.2.BD$OTU = "OTU.1"

binwidth = 0.05

p = ggplot(OTU.2.BD, aes(x = BD))

p = p + geom_histogram(alpha = 0.5, binwidth = binwidth, aes(y = ..density..))

p
```

```python
%%R
binwidth = 0.05
bins = seq(0,1,binwidth)
OTU.1.hist = hist(OTU.1.BD$BD, bins = bins, plot = FALSE)
OTU.2.hist = hist(OTU.2.BD$BD, bins = bins, plot = FALSE)
```

##Large sigma

```python
%%R
C = function(mids, density, x = seq(0,1,0.01), binwidth = 0.05, sigma = 0.10) {
    #area = c0 * sigma * sqrt(2 * pi)
    area = density * binwidth 
    c0 = area / (sigma * sqrt(2 * pi))
    c0 * exp(-(mids - x)**2 / sigma ** 2)
}
```

```python
%%R -h 300 -w 600
binwidth = 0.05
bins = seq(0,1,binwidth)
OTU.1.hist = hist(OTU.1.BD$BD, bins = bins, plot = FALSE)

d = with(OTU.1.hist, data.frame(density = density, mids = mids))
d$value = "hist"

d.tmp = data.frame(mids = seq(0,1,0.01), density = rowSums(mapply(C, d$mids, d$density)))
d.tmp$value = "gauss"

d = rbind(d, d.tmp)

p = ggplot(d, aes(x = mids, y = density, color = value, fill = value))

p = p + geom_bar(data = subset(d, value == "hist"), stat = "identity", alpha = 0.50)

p = p + geom_line(data = subset(d, value == "gauss"), size = 2)

p = p + labs(y = "Concentration", x = "Density")

p
```

```python
%%R -h 300 -w 600
binwidth = 0.05
bins = seq(0,1,binwidth)
OTU.2.hist = hist(OTU.2.BD$BD, bins = bins, plot = FALSE)

d = with(OTU.2.hist, data.frame(density = density, mids = mids))
d$value = "hist"

d.tmp = data.frame(mids = seq(0,1,0.01), density = rowSums(mapply(C, d$mids, d$density)))
d.tmp$value = "gauss"

d = rbind(d, d.tmp)

p = ggplot(d, aes(x = mids, y = density, color = value, fill = value))

p = p + geom_bar(data = subset(d, value == "hist"), stat = "identity", alpha = 0.50)

p = p + geom_line(data = subset(d, value == "gauss"), size = 2)

p = p + labs(y = "Concentration", x = "Density")

p
```

```python
%%R -w 800 -h 300
f = function(hist) {
    d = with(hist, data.frame(concentration = density, density = mids))
    d$value = "hist"
    d.tmp = data.frame(density = seq(0,1,0.01), concentration = rowSums(mapply(C, d$density, d$concentration)))
    d.tmp$value = "gauss"
    d = rbind(d, d.tmp)
    d$name = hist$xname
    d
}

df = ldply(list(OTU.1.hist, OTU.2.hist), f)

p = ggplot(df, aes(x = density, y = concentration, color = value, fill = value))

p = p + facet_wrap(~name)

p = p + geom_bar(data = subset(df, value == "hist"), stat = "identity", alpha = 0.50)

p = p + geom_line(data = subset(df, value == "gauss"), size = 2)

p = p + labs(y = "Concentration", x = "Density")

p
```

```python
%%R -w 700 -h 300
d = df %>% 
    filter(value == "gauss") %>%
    group_by(density) %>%
    mutate(total = sum(concentration)) %>%
    group_by(density, name) %>%
    mutate(conc.norm = concentration / total) %>%
    group_by() %>%
    select(conc.norm, name, density)

p = ggplot(d, aes(x = density, y = conc.norm, fill = name))

p = p + geom_area()

p
```
