##These first few cells make the dummy data for the simulation below. In practice, all of this hypthetical data could be substituted for real numbers (e.g. actual BD histograms or actual rank abundance charts).

```python
%%R -w 300 -h 250
N.otus = 1000
abund.df = data.frame(Abundances = rlnorm(N.otus, sdlog = 2))
abund.df$Rank = rank(-abund.df$Abundances)
abund.df$OTU = paste("OTU.", abund.df$Rank, sep = "")

p = ggplot(abund.df, aes(x = Rank, y = Abundances))

p = p + geom_line()

p = p + scale_y_log10()

p
```

```python
%%R
N = 1000
binwidth = 0.05
bins = seq(0,1,binwidth)

get.BD.hist = function ( abundance ) {
    mu = runif(1, 0.40, 0.60)
    alpha = runif(1, 0.15, 0.35)
    x = rnorm(N, mean = mu, sd = alpha)
    BD.hist = hist(x[(x > 0)&(x < 1)], bins = bins, plot = FALSE)
    with(BD.hist, data.frame(hist.density = density * abundance, mids = mids))
}
```

```python
%%R
abundances = abund.df$Abundances
names(abundances) = abund.df$OTU
hist.df = ldply(abundances, get.BD.hist, .id = "OTU")
```

###Here are 100 randomly selected BD profiles (out of 100 totals) that we'll be working with...

```python
%%R -w 800 -h 800
otus = base::sample(hist.df$OTU, 100)
p = ggplot(hist.df %>% filter(OTU %in% otus), aes(x = mids, y = hist.density))
p = p + facet_wrap(~OTU, ncol = 10, scales = "free_y")
p = p + geom_line()

p = p + theme_bw()
p = p + theme(axis.text = element_text(size = 6))

p
```

##This is where the model would begin if you wanted to use actual data. At the moment all values are in relative gradient length units. That is, the length of the gradient is set to 1 (so a density of 0.10 in cells below corresponds to 1/10 the gradient length). Substituting in actual densities with the relative units is trivial.

```python
%%R
X = seq(0,1,0.01)

C.bin = function(mids, hist.density, x = X, binwidth = 0.05, sigma = 0.10) {
    #area = c0 * sigma * sqrt(2 * pi)
    area = hist.density * binwidth 
    c0 = area / (sigma * sqrt(2 * pi))
    c0 * exp(-(mids - x)**2 / sigma ** 2)
}

C.gradient = function(mids, hist.density) {
    data.frame(Conc = rowSums(mapply(C.bin, mids, hist.density)), 
               Density = X)
}
```

```python
%%R
OTU.conc.df = hist.df %>%
    group_by(OTU) %>%
    do(C.gradient(mids = .$mids, hist.density = .$hist.density))

OTU.conc.df = left_join(OTU.conc.df, abund.df %>% select(OTU, Rank))
```

###Here is the total DNA curve across the gradient based on the modelled concetrations and BD histograms (above).

```python
%%R -w 350 -h 250
d = OTU.conc.df %>% 
    group_by(Density) %>%
    mutate(Total = sum(Conc))

p = ggplot(d, aes(x = Density, y = Total))

p = p + geom_line()

p
```

###Here are individual OTU concentration profiles (concentrations are the sum of guassians centered at each bin in the BD histogram. So, the concentration profile for each OTU, in this example, is the sum of twenty gaussians that follow the BD histogram).

```python
%%R -w 600 -h 500
other = 15
OTU.conc.df$to.plot = with(OTU.conc.df, ifelse(Rank <= other, OTU, "other"))
OTU.conc.df

d = OTU.conc.df %>% 
    group_by(Density) %>%
    mutate(Total = sum(Conc)) %>%
    group_by(Density, to.plot) %>%
    summarize(Conc.norm = first(sum(Conc) / Total),
              Conc = sum(Conc)) %>%
    group_by() %>%
    gather(key, value, Conc.norm:Conc)

p = ggplot(d, aes(x = Density, y = value, fill = to.plot))

p = p + facet_wrap(~key, scales = "free_y", ncol = 1)

p = p + geom_area()

p
```
